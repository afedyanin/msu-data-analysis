```python
# импортируем необходимые библиотеки
import numpy as np

# создаем массивы значений
x = np.random.randn(10000000)
y = np.random.randn(10000000)
```

```python
%%time

nx = len(x)
result = 0.0
count = 0
for i in range(nx):
    result += x[i] - y[i]
    count += 1
result / count
```

```python
%%timeit

nx = len(x)
result = 0.0
count = 0
for i in range(nx):
    result += x[i] - y[i]
    count += 1
result / count
```
